```python
import pandas as pd
import shutil
import os

train_path = '/home/projects/vokquant/accent-recog-slt2022/CommonAccent/data/at_augmented_dev_regions/train_augmented_regions_offset.csv'
valid_path = '/home/projects/vokquant/accent-recog-slt2022/CommonAccent/data/at_augmented_dev_regions/dev_augmented_regions_offset.csv'
test_path = '/home/projects/vokquant/accent-recog-slt2022/CommonAccent/data/at_augmented_dev_regions/test_augmented_regions_offset.csv'

train_df = pd.read_csv(train_path)
valid_df = pd.read_csv(valid_path)
test_df = pd.read_csv(test_path)

def get_wav_list(df):
    wav_list = []
    for i in range(len(df)):
        wav_list.append(df.iloc[i]['wav'])
    return wav_list

# get all wavs
train_wav_list = get_wav_list(train_df)
valid_wav_list = get_wav_list(valid_df)
test_wav_list = get_wav_list(test_df)

# combine all wavs of train, valid, test to a zip file
all_wav_list = train_wav_list + valid_wav_list + test_wav_list
print(len(all_wav_list))

save_dir = '/home/projects/vokquant/data/dicla/augmented_dev'
os.makedirs(save_dir, exist_ok=True)

# copy all wavs to zip file
copy_wav = False
if copy_wav:
    for wav in all_wav_list:
        shutil.copy(wav, save_dir)
    
# make tar file from files in save_dir

os.system(f'tar -cvf {save_dir}.tar {save_dir}')
# better:
# os.system(f'cd {os.path.dirname(save_dir)} && tar -cvf {os.path.basename(save_dir)}.tar {os.path.basename(save_dir)}')
```
